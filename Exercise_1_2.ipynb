**Exercise 4:** Rewrite the grade program from the previous chapter using a function called computegrade that takes a score as its parameter and returns a grade as a string.

```python
def computegrade(score):
  if score < 0.0 or score > 1.0:
    return "Bad score"
  elif score >= 0.9:
    return "A"
  elif score >= 0.8:
    return "B"
  elif score >= 0.7:
    return "C"
  elif score >= 0.6:
    return "D"
  else:
    return "F"

while True:
  score_input = input("Enter score (or 'exit' to quit): ")
  if score_input.lower() == 'exit':
    break
  try:
    score = float(score_input)
    print(computegrade(score))
  except ValueError:
    print("Bad score")
```

**Exercise 1:** Write a program which repeatedly reads integers until the user enters “done”. Once “done” is entered, print out the total, count, and average of the integers. If the user enters anything other than a integers, detect their mistake using try and except and print an error message and skip to the next integers.

```python
total = 0
count = 0

while True:
  num_input = input("Enter a number: ")
  if num_input == "done":
    break

  try:
    num = int(num_input)
    total += num
    count += 1
  except:
    print("Invalid input")

if count > 0:
  average = total/count
  print(total, count, average)
else:
  print("No valid input")
```

**Exercise 2:** Write another program that prompts for a list of numbers as above and at the end prints out both the maximum and minimum of the numbers instead of the average.

```python
numbers = []

while True:
  num_input = input("Enter a number: ").strip()

  if num_input.lower() == "done":
    break

  try:
    num = int(num_input)
    numbers.append(num)
  except ValueError:
    print("Invalid input")

if numbers:
  print("Maximum:", max(numbers))
  print("Minimum:", min(numbers))
else:
  print("No valid numbers were enetered")
```
