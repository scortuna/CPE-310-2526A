**Part 1:**

Suppose we had a linear regression model with one feature that learned the following bias and weights:

```python
## GIven
b = 1
w1= 3
x1= 0

y = w1*x1 + b
print(f'y = {y}')

y = 1
import matplotlib.pyplot as plt
w1 = 3
b= 1

x = [-5, 5]
y= [w1 * i + b for i in x]
x1 = 0
y1= w1*x1 + b

plt.plot(x,y)
plt.scatter(x1, y1)

plt.show()
```

**Analysis**: In this linear regression model, the input feature $x_1 = 0$ nullifies the weighted term $w_1 \cdot x_1$, causing the prediction $y$ to align perfectly with the model's bias $b = 1$. This result sits exactly on the y-intercept of the plotted line, confirming that when the feature is zero, the prediction is defined entirely by the baseline bias.
